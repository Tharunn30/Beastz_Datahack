```python
tags = video.sort_values('view_count',ascending=False).iloc[:,7].head().to_list()
distances = np.zeros((len(tags), len(tags)))
for i in range(len(tags)):
    for j in range(len(tags)):
        distances[i, j] = Levenshtein.distance(tags[i], tags[j])

# Find similar tags
similar_tags = set()
for i in range(len(tags)):
    for j in range(i+1, len(tags)):
        if distances[i, j] <= 2:
            similar_tags.add(tags[i])
            similar_tags.add(tags[j])
print("All the similar tags of the videos that did well")
similar_tags
print("All this is from the same channel: \n",)
```
